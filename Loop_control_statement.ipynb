### *Loop Control Statements*

#### Loop control statements change the execution of the program from its normal sequence
####  1.Break    2.Continue      3.Pass 

```python
# 1. Break Statement
# The break statement is used to terminate the loop prematurely, regardless of the loop's condition.
```

```python
for i in range(1,10):
    if i == 7:
        break
    print(i)
```

```python
string="Infosys"
for letter in string:
    if letter=="y" or letter=="s":
        break
    print(letter)
print("Out of the loop")
```

```python
str="python"
for i in str:
    if (i=="t"):
        print("t found")
        break
    print(i)
else:
    print("END") # In Python, the else block in a for loop runs only if the loop completes normally,i.e., without hitting a break. 
```

```python
str = "python"
for i in str:
    if i == "z":   # 'z' is not in string
        print("z found")
        break
    print(i)
else:
    print("END") # in this case else block will execute as loops completes normally
```

```python
# Ask the user for a password up to 3 times;stop early if found
```

```python
correct_pwd="pass1234"
count=3
for i in range(count):
    password=input("Enter the password:")
    if password==correct_pwd: 
        print("Correct password")    
        break
    else:
        print("Incorrect password")
else:
    print("too many attempts")        
```

```python
# 2. continue Statement
# The continue statement is used to skip the current iteration and move to the next iteration of the loop.
```

```python
# Print the number from 1 to 10
```

```python
for i in range(1,11):
    if i==5:
        continue
    print(i)
```

```python
# Print odd numbers from 1 to 10
```

```python
for i in range(1,11):
    if i%2==0:
        continue
    print(i)
```

```python
# From [3,7,1,8,4,9] print only numbers>=5
```

```python
nums=[3,7,1,8,4,9]
for n in nums:
    if n<5:
        continue
    print(n)
```

```python
# From ["apple", "", "banana", "", "mango"], print only non-empty strings
```

```python
fruits=["apple", "", "banana", "", "mango"]
for fruit in fruits:
    if fruit==" ":
        continue
    print(fruit)
```

```python
# Print numbers from a list, skip zeros, and stop when you find a 5.
```

```python
nums=[0, 0, 2, 3, 5, 7]
for num in nums:
    if num==0:
        continue
    if num==5:
        break
    print(num)
```

```python
# Take range from user i.e. upper and lower limit from user and print addition of only even numbers between the range.
```

```python
lower=int(input("Enter the number:"))
upper=int(input("Enter the number:"))
sum=0
for i in range(lower,upper+1):
    if i%2!=0:
        continue
    sum+=i
print(sum)
```

```python
# 3.pass Statement
# The pass statement is a placeholder that does nothing. It is often used as a placeholder for future code or to handle syntactical requirements.
```

```python
for num in range(1, 5):
    if num == 3: 
        pass
    print(num)
```
