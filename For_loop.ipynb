# *For Loop*

#### for variable in sequence:
    # loop body
    
variable → takes each value from the sequence (one by one).
sequence → can be a list, tuple, string, range, etc.
Loop body → code inside will run for each item

- For traversing(visit each element one by one) the data type, a for loop is used

- Used when we know the number of iterations in advance (e.g., iterating over a range, list, string, etc.).

- It automatically takes each element from a sequence one by one

# *Pattern Programs* 

```python
# Right angled triangle
```

```python
rows=int(input("Enter the number of rows:"))
for i in range(1,rows+1):
    print("* " * i)  
```

```python
rows=int(input("Enter the number of rows:"))
for i in range(1,rows+1):       # outer loop controls the row number
    for j in range(1,i+1):      # inner loop prints stars in each row
        print("*",end=" ")      # prints star without moving to next line
    print()                     # moves to next line after each row
```

```python
# Number Pyramid
```

```python
rows=int(input("Enter the number of rows:"))
for i in range(0,rows+1):
    for j in range(0,rows-i): # Prints leading spaces
        print(end=" ")
    for j in range(1,i+1):    # Prints numbers starting from 1 up to i
        print(j,end=" ")
    print()
```

```python
# Star Pyramid
```

```python
rows=int(input("Enter the number of rows:"))
for i in range(0,rows):
    for j in range(0,rows-i-1):
        print(end=" ")
    for j in range(0,i+1):
        print("*",end=" ")
    print()
```

```python
# Star Pyramid (odd number of stars)
```

```python
rows=int(input("Enter the number of rows:"))
for i in range(0,rows):
    for j in range(0,rows-i-1):
        print(end=" ")
    for j in range(0,2*i+1):
        print("*",end=" ")
    print()
```

```python
# Inverted Pyramid
```

```python
rows = int(input("Enter the number of rows: "))
for i in range(rows,0,-1):
    for j in range(0,rows-i):           
        print(end=" ")
    for j in range(0,i):       
        print("*", end=" ")
    print()
```

```python
# Diamond Pattern
```

```python
rows=int(input("Enter the number of rows:"))
for i in range(0,rows):
    for j in range(0,rows-i-1):
        print(end=" ")
    for j in range(0,i+1):
        print("*",end=" ")
    print()
for i in range(0, rows):
    for j in range(0, i):           
        print(end=" ")
    for j in range(0, rows - i):       
        print("*", end=" ")
    print()    
```

```python
# Floyd’s Triangle
```

```python
rows=int(input("Enter the total Number of Rows:"))
number=1
print("Floyd's Triangle")
for i in range(1,rows+1):
   for j in range(1,i+1):
       print(number,end=" ")
       number+=1
   print()
```

```python
# Pascal’s Triangle
```

```python
n = int(input("Enter number of rows: "))
for i in range(n):
    num = 1
    for j in range(i + 1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)
    print()
```
