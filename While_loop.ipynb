# *While loop*

#### while condition:
    # loop body
    # update statement
Explanation:
Initialisation → set a starting value (before the loop).
Condition → loop keeps running as long as this is True.
Loop body → code that executes repeatedly.
Update statement → change the loop variable, so eventually the condition becomes False (otherwise → infinite loop).

- Used when the number of iterations is not fixed and depends on a condition.

- It keeps running as long as the condition is True.

```python
# Reverse a number using a while loop
```

```python
num=int(input("Enter the number:"))
rev=0
while num>0:
    digit=num%10
    rev=rev*10 + digit
    num//=10
print("Reverse number is",rev)
```

```python
# Check Palindrome Number
```

```python
n=int(input("Enter a number: "))
original=n
rev=0
while n>0:
    digit=n%10
    rev=rev*10 + digit
    n//=10
if original == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Armstrong number
```

```python
num=int(input("Enter a number: "))
sum=0
temp=num
while num > 0:
    digit=num % 10
    sum+=digit ** 3  
    num//=10
if sum==temp:
    print(temp,"is an Armstrong number")
else:
    print(temp,"is not an Armstrong number")
```

```python
# Check whether a number is prime or not
```

```python
n=int(input("Enter the number:"))
i=1
count=0
while i<=n:
    if n%i==0:
        count+=1
    i+=1
print(f"Total factors of {n} are",count)
if count==2:
    print(f"{n} is prime number")
else:
    print(f"{n} is not a prime number")
```

```python
# Find largest and smallest digit in a number
```

```python
n=int(input("Enter a number: "))
largest=0
smallest=9
while n>0:
    digit=n%10
    if digit>largest:
        largest=digit
    if digit<smallest:
        smallest=digit
    n//=10
print("Largest digit:",largest)
print("Smallest digit:",smallest)
```

```python
# Guessing game
```

```python
import random
jackpot=random.randint(1,100)
guess=int(input("Enter to guess the number:"))
counter=1
while guess!=jackpot:
    if guess<jackpot:
        print("Wrong!...Guess higher")
    else:
        print("Wrong!...Guess Lower")
        
    counter+=1
    guess=int(input(f"We are giving you {counter}nd chance...Enter the number again:"))

print("Correct guess")
print("You have guess the number correctly in attempts:",counter)
```

```python
# Guessing game with attempt limit
```

```python
import random
jackpot=random.randint(1, 100)
guess=int(input("Enter to guess the number (1–100):"))
counter=1
max_attempts=3

while guess!=jackpot and counter<max_attempts:
    if guess<jackpot:
        print("Wrong!...Guess higher")
    else:
        print("Wrong!...Guess lower")
    
    counter+=1
    guess=int(input(f"This is your {counter} attempt...Enter the number again:"))

if guess==jackpot:
    print("Correct guess!")
    print("You guessed the number correctly in attempts:",counter)
else:
    print("Game Over! You have used all attempts.")
    print("The correct number was:",jackpot)
```
