```python
# Triangle Type
```

```python
A=int(input("Enter the first side of triangle"))
B=int(input("Enter the second of triangle"))
C=int(input("Enter the third side of triangle"))
if A+B>C and A+C>B and B+C>A:
    if A == B == C:
        print("Equilateral Triangle")
    elif A == B or B == C or A == C:
        print("Isosceles Triangle")
    else:
        print("Scalene Triangle")
else:
    print("Not a valid triangle")
```

```python
menu = print ("""---Welcome to Calculator Program----\n
1.Enter 1 for Addition(+)
2.Enter 2 for Subtraction(-)
3.Enter 3 for Multiplication(*)
4.Enter 4 for Division(/)
5.Enter 5 for Floor Division(//)
""")

num1=int(input("Enter the first number:"))
num2=int(input("Enter the second number:"))
ch=input("Enter Your Choice to perform Arithmetic Operation(1,2,3,4,5):")

if ch=="1":
    print("Addition of two number is",num1+num2)
elif ch=="2":
    print("Subtraction of two number is",num1-num2)
elif ch=="3":
    print("Multiplication of two number is",num1*num2)
elif ch=="4":
    if num2!=0:
        print("Division of two number is",num1/num2)
    else:
        print("Cannot perform Division by zero")
elif ch=="5":
    if num2!=0:
        print("Floor Division of two number is",num1//num2)
    else:
        print("Cannot perform floor Division by zero")   
else:
    print("Exit")
```
