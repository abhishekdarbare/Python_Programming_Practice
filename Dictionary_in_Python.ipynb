## Dictionary

#### If we want to represent a group of objects as key-value pairs then we should go for Dictionary.

#### Duplicate keys are not allowed but values can be duplicated.

#### Hetrogeneous objects are allowed for both key and values.

#### Dictionaries are mutable and dynamic

#### indexing and slicing concepts are not applicable

#### syntax            dict = { key : 'value' , key : 'value' }

## Characterstics:
#### 1) Mutable 2) Indexing has no meaning 3) keys can't be duplicated
#### 4) keys can't be mutable items

## Create Dictionary

```python
# empty dictionary
d = {}
print(d)
print(type(d))
```

```python
# 1D dictionary --> Homogeneous
d1 = { 'name' : 'Ram' ,'gender' : 'male' }
print(d1)
```

```python
# with mixed keys --> Heterogeneous
d2 = {(1,2,3):1,'hello':'world'}
print(d2)
```

```python
# 2D dictionary
s = {'name':'Ram','college':'mit','sem':4,
     'subjects':{'dsa':50,'maths':67,'english':34}
     }
print(s)
```

### using sequence and dict function

```python
d4 = dict([('name','Ram'),('age',32),('Roll_No',11)])
print(d4)
```

```python
# duplicate keys
d5 = {'name':'Ram','name':'Ganesh'}
print(d5)
```

#### keys can't be mutable items -> i.e List,Set

```python
d6 = {'name':'Ram',{1,2,3}:2}
print(d6) #TypeError:unhashable type: 'set'
```

## Accessing items

```python
my_dict={'name':'Jack','age':26,'Course':'DA','city':'Mumbai'}
print(my_dict['Course'])
# print(my_dict['email']) #KeyError: 'email'-->email as a key is not present
```

### get()
#### Returns the value for the specified key,or None if key is not found
#### dict.get(key)

```python
my_dict={'name':'Jack','age':26,'Course':'DA','city':'Mumbai'}
print(my_dict.get('city'))
print(my_dict.get('email'))
```

#### for 2D Dictionary

```python
s={'name':'Ram','college':'mit','sem':4,
     'subjects':{'dsa':50,'maths':67,'english':34}
     }
print(s)
print(s['subjects'])
print(s['subjects']['maths']) #2D access
print(s.get('subjects'))
print(s.get('subjects').get('maths')) #2D access
```

### Dictionary Operations  - Membership

```python
s = {'name':'Ram','college':'mit','sem':4,
     'subjects':{'dsa':50,'maths':67,'english':34}
     }
print('name' in s)
```

```python
print('maths' in s)
print('maths' in s['subjects'])
print('subjects' in s)
```

```python
#Iteration
d = {'name':'Ram','gender':'male','age':33,'city':"pune"}
for i in d:
  print(i,":",d[i])
```

### Remove key-value pair --> you can pass only the key that you want to delete. You can't pass the value
#### 1. pop()     2.popitem()      3.del()        4.clear()

```python
# 1.pop --> This function accept key and removes that key:value pair
```

```python
d = {'name': 'Ram', 'age': 32, 3: 113, 'gender': 'male', 'weight': 72}
d.pop('weight')
print(d)
```

```python
# 2.popitem --> This fn has no argument.It deletes last key:value pair by default
```

```python
d = {'name': 'Ram', 'age': 32, 3: 113, 'gender': 'male', 'weight': 72}
d.popitem()
print(d)
```

```python
# 3.del--> This fn.accept key and removes that key:value pair.It can deletes entire dictionary
```

```python
d = {'name': 'Ram', 'age': 32, 3: 113, 'gender': 'male', 'weight': 72}
#del d['name']
del d
print(d) #NameError: name 'd' is not defined
```

```python
# 4.clear --> This function deletes all items from dictionary.
d = {'name': 'Ram', 'age': 32, 3: 113, 'gender': 'male', 'weight': 72}
d.clear()
print(d)
```

## Dictionary Methods

```python
d={"name":"Ram","Email":"fusion@gmail.com",
      "location":"pune","Mob_no":1234567890,
      'gender':'male','age':33}
print(d)
```

```python
#len/sorted/min/max
print(len(d))
print(sorted(d,reverse=True))
print(max(d))
print(min(d))

d1={10:"Hi",20:"hello",30:"Python"}
print(sum(d1))
print(max(d1))
print(min(d1))
```

####  1.keys()
Returns a view object with a list of dictionary's keys.
dict.keys()

```python
d={"name":"Ram","Email":"fusion@gmail.com",
      "location":"pune","Mob_no":1234567890,
      'gender':'male','age':33}
print(d.keys())
```

#### 2.values
dict.values()
Returns a view object with a list of dictionary's values.

```python
d={"name":"Ram","Email":"fusion@gmail.com",
      "location":"pune","Mob_no":1234567890,
      'gender':'male','age':33}
print(d.values())
```

#### 3.items()
Returns a view object with a list of dictionary's key-value pairs.
dict.items()

```python
d={"name":"Ram","Email":"fusion@gmail.com",
      "location":"pune","Mob_no":1234567890,
      'gender':'male','age':33}
print(d.items())
```

#### update() --> Update function permanently makes the changes in first i/p
#### Updates the dictionary with another dictionary or iterable of key-value pairs.
##### dict.update([other])

```python
d1 = {'Name': 'Ram', 'Age': '19', 'Country': 'India'}
d2 = {'Name': 'Neha', 'Age': '22'}

d1.update(d2)
print(d1)
```
