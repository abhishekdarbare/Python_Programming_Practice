### Tuple()

* Tuple is exactly same as List except that it is immutable. i.e once we creates Tuple object,we cannot perform any changes in that object. Hence Tuple is Read Only version of List.
* If our data is fixed and never changes then we should go for Tuple.
* Insertion Order is preserved i.e. ordered collection
* Duplicates are allowed
* Heterogeneous objects are allowed.
* We can preserve insertion order and we can differentiate duplicate objects by using index. Hence index will play very important role in Tuple also

```python
# create a tuple with a single item
t1=('hello')
t2=('hello',)
print(type(t1))
print(type(t2))
```

```python
# homo
t3 = (1,2,3,4)
print(t3)
# hetro
t4 = (1,2.5,True,[1,2,3])
print(t4)
```

```python
# tuple -->2D Tuple
t5 = (1,2,3,(4,5))
print(t5)
```

```python
# using type conversion
t6 = tuple([1,2,3])
print(t6)
print(type(t6))
```

### Accessing Items --> Indexing Slicing

```python
t3 = (1,2,3,4)
print(t3)
print(t3[0])
print(t3[-1])
print(t3[ : :-1])
```

```python
t5 = (1,2,3,(4,5))
print(t5[-1][0])
```

```python
t = (1,2,3,4,5)
print(t[-2:-5:-2])
```

 ### Editing items -> immutable just like strings.Editing items and adding items  in Tuple not possible.

```python
t3 = (1,2,3,4)
t3[0] = 100
print(t3)

#This code will give us the error
#TypeError: 'tuple' object does not support item assignment
```

### Deleting items -> items in tuple can't be deleted but entire tuple can be deleted using del function

```python
t3 = (1,2,3,4)
print(t3)
del t3  #Entire tuple is deleted
print(t3)
```

```python
t5 = (1,2,3,(4,5))
print(t5)
del t5[-1] #deleting a single element in a tuple is not possible
print(t5)

#This code will give us the error
#TypeError: 'tuple' object doesn't support item deletion
```

### Difference between Lists and Tuples
##### 1] Syntax
##### 2] Mutability
##### 3] Speed:- In terms of execution time, Tuples are faster than Lists
##### 4] Memory:- A List occupies more memory than a Tuple.
##### 5] Built-in functionality

```python
import sys
L = list(range(0,1000))
T = tuple(range(0,1000))

print('List size',sys.getsizeof(L))
print('Tuple size',sys.getsizeof(T))
```

### Zipping tuples
#The zip() function returns a zip object, which is an iterator of tuples
#### where the first item in each passed iterator is paired together, and then the
#### The second item in each passed iterator is paired together.
#If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
a = (1,2,3,4)
b = (5,6,7,8)

print(list(zip(a,b)))
print(tuple(zip(a,b)))
```

```python
# Swapping
a = 1
b = 2
a,b = b,a
print(a,b)
```

### *List Comprehension Practise questions*

```python
# 1.print the list of tuples(x,y)i.e unique paies from the list
```

```python
for x in range(0,3):
    for y in range(0,2):
        print(x,y)
```

```python
# using List Comprehension
result=[(x, y) for x in range(0,3) for y in range(0,2)]
print(result,end="")
```

```python
# 2. Print the sum of two lists using list comprehension
```

```python
a=[1,2,3]
b=[4,5,6]
# result=[x+y for x in a for y in b]
# print(result)
print([x+y for x,y in zip(a,b)]) #index wise addition using tuple zip function
```

```python
# 3.Using Zip combine two lists
```

```python
names =["Alice", "Bob", "Charlie"]
scores=[85, 90, 95]
result=[(name, score) for name,score in zip(names, scores)] 
print(result)
```
