### Aggregation(Has-A relationship)

```python
# Aggregation in object-oriented programming (OOP) is a type of association that represents a "has-a" relationship between classes,
# where one class contains a reference to another class but both can exist independently.
# It's a way to model relationships where the child can exist without the parent
```

```python
# Here, Course objects are created independently and then associated with the Student object.
# If john is deleted, math and science still exist.

class Course:
    def __init__(self, name):
        self.name = name

class Student:
    def __init__(self, name):
        self.name = name
        self.courses = []  # Aggregation: Student has a list of Course objects

    def enroll(self, course):
        self.courses.append(course)

# Usage
math = Course("Mathematics")
science = Course("Science")

john = Student("John")
john.enroll(math)
john.enroll(science)

for course in john.courses:
    print(course.name)
```

```python
# example
class Customer:

  def __init__(self,name,gender,address):
    self.name = name
    self.gender = gender
    self.address = address

  def print_address(self):
    print(self.address._Address__city,self.address.pin,self.address.state)

  def edit_profile(self,new_name,new_city,new_pin,new_state):
    self.name = new_name
    self.address.edit_address(new_city,new_pin,new_state)

class Address:

  def __init__(self,city,pin,state):
      self.__city = city
      self.pin = pin
      self.state = state

  def get_city(self):
    return self.__city

  def edit_address(self,new_city,new_pin,new_state):
    self.__city = new_city
    self.pin = new_pin
    self.state = new_state

add1 = Address('Indore',122011,'Madhya Pradesh')
cust = Customer('Abhishek','Male',add1)

print(cust.print_address())

cust.edit_profile('Aryan','Mumbai',144108,'Maharashtra')
print(cust.print_address())
```

### Inheritance

- What is inheritance
- Example
- What gets inherited?

```python
# Inheritance enables a class (called the child class) to inherit attributes and methods from another class (called the parent class).
# promoting code reuse and logical hierarchy
```

Key Benefits of Inheritance
- Code Reusability: Common functionality in the parent class can be reused by child classes, reducing duplication.
- Modularity: Logical grouping of shared behavior makes code easier to manage and understand.
- Extensibility: New features can be added to child classes without modifying the parent class.
- Polymorphism: Enables objects of different subclasses to be treated as instances of the parent class, allowing flexible and dynamic code.
- Maintainability: Changes in the parent class automatically propagate to child classes, simplifying updates and bug fixes


```python
# Example

# parent
class User:

  def __init__(self):
    self.name = 'Abhishek'
    self.gender = 'Male'

  def login(self):
    print('login')

# child
class Student(User):

  # def __init__(self):
  #   self.rollno = 100

  def enroll(self):
    print('enroll into the course')

u = User()
s = Student()

print(s.name)
print(s.gender)
s.login()
s.enroll()
```

### Super Keyword

```python
# The super keyword in Python is used to call methods or constructors from a parent class.
# It's especially useful in inheritance when you want to extend or customise behaviour without completely overriding it.
```

```python
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print ("Buying a smartphone")
        # syntax to call parent ka buy method
        super().buy()

s=SmartPhone(20000, "Apple", 13)

s.buy()
```

```python
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print ("Buying a smartphone")
        # syntax to call parent ka buy method
        print(super().brand)
# super() lets you call parent class methods and constructors, but not instance variables directly.
# To access parent data, use self.attribute or wrap it in a method like get_brand().

s=SmartPhone(20000, "Apple", 13)

s.buy()
```

```python
# super -> constuctor
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

class SmartPhone(Phone):
    def __init__(self, price, brand, camera, os, ram):
        print('Inside smartphone constructor')
        super().__init__(price, brand, camera)
        self.os = os
        self.ram = ram
        print ("Inside smartphone constructor")

s=SmartPhone(20000, "Samsung", 12, "Android", 2)

print(s.os)
print(s.brand)
```

```python
# super cannot access variables
# super cannot be used outside the class
# super is always used inside the child class
```

### Polymorphism

- Method Overriding
- Method Overloading
- Operator Overloading

Polymorphism means "many forms". It refers to the ability of an entity (like a function or object) to perform different actions based on the context.

##### Why do we need Polymorphism?
- Ensures consistent interfaces across different classes.
- Allows objects to respond differently to the same method call.
- Promotes loose coupling by relying on shared behaviour, not specific types.
- Enables writing flexible, reusable code that works across types.
- Simplifies testing and future extension of code.

```python
# Method Overloading(does not work in Python)
class Shape:
    def area(self,radius):
        return 3.14*radius*radius

    def area(self,l,b):
        return l*b
        
s = Shape()
s.area(2)
s.area(3,4)
```

```python
# Method overloading is the ability to define multiple methods with the same name but different parameters within the same class. 
# Python does not support traditional method overloading because it is dynamically typed and treats function names as variables, 
# so redefining a function overwrites the previous one. 
# Instead, Python achieves similar behaviour using default arguments and variable-length arguments (*args).

class Shape:
    def area(self,a,b=0):
        if b==0:
            return 3.14*a*a
        else:
            return a*b
        
s = Shape()
print(s.area(2))
print(s.area(3,4))
```

#### Types of Polymorphism
1. Compile-time Polymorphism
2. Runtime Polymorphism (Overriding)

```python
# compile-time
class Calculator:
    def multiply(self, a=1, b=1, *args):
        result = a * b
        for num in args:
            result *= num
        return result

# Create object
calc = Calculator()

# Using default arguments
print(calc.multiply())            
print(calc.multiply(4))           

# Using multiple arguments
print(calc.multiply(2, 3))       
print(calc.multiply(2, 3, 4))
```

```python
# run-time
class Animal:
    def sound(self):
        return "Some generic sound"

class Dog(Animal):
    def sound(self):
        return "Bark"

class Cat(Animal):
    def sound(self):
        return "Meow"

# Polymorphic behaviour
animals = [Dog(), Cat(), Animal()]
for animal in animals:
    print(animal.sound())
```

### Abstraction

Abstraction means hiding the complex implementation details and showing only the essential features of an object or system.
In Simple words, you interact with a simple interface while the complex mechanics are hidden.

```python
from abc import ABC,abstractmethod
class BankApp:

    def database(self):
        print("Connected to Database")

    @abstractmethod
    def security(self):
        pass

    @abstractmethod
    def display(self):
        pass    
```

```python
class MobileApp(BankApp):

    def mobile_login(self):
        print("login into mobile")

    def security(self):
        print("mobile security")

    def display(self):
        print("display")
```
