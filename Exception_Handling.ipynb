### Catching Specific Exceptions

```python
#Ex1
a=int(input("Enter Numerator:"))
b=int(input("Enter Denominator:"))
print(a/b)
```

```python
#Ex1-Exception Handling
try:
    a=int(input("Enter Numerator:"))
    b=int(input("Enter Denominator:"))
    print(a/b)
except ZeroDivisionError:
    print("cannot divide by zero")
except ValueError:
    print("pls enter numeric value only")
```

```python
#Ex2
data=["10","20","thirty","40"]
numeric_data=[]
for i in data:
    try:
        numeric_data.append(float(i))
    except ValueError:
        print(f"Skipping invalid value:{i}")
print("Cleaned data:",numeric_data)
```

```python
#Ex3
def fun(a):
    if a<4:
        b=a/(a-3)
    print("Value of b=",b)
try:
    fun(5)
    #fun(3)
    #fun("hi")
except ZeroDivisionError:
    print("ZeroDivisionError Occurred and Handled")
except NameError:
    print("NameError Occurred and Handled")
except TypeError:
    print("You have to pass only int value not string")
```

#### Using else and finally Blocks

```python
def my_function(a,b):
    try:
        c=((a+b) // (a-b))
    except ZeroDivisionError:
        print("Denominator cannot be zero")
    else:
        print(c)

my_function(2,3)
#my_function(3,3)
```

```python
import pandas as pd
try:
    df=pd.read_csv("sales_data.csv")
except FileNotFoundError:
    print("The file does not exist,pls provide the valid path of file ")
except Exception as e:
    print("The file is empty.")
else:
    print(df.head())
finally:
    print("File loading process completed.")
```

```python
#Handling multiple exceptions in one line
try:
    a=int(input("Enter number:"))
    b=int(input("Enter divisor:"))
    c=((a+b) /(a-b))
except (ValueError, ZeroDivisionError, TypeError) as e:
    print("Error:",e)
else:
    print(c)
finally:
    print("This is always excecuted")
```

### raise keyword

The raise keyword is used to manually generate (raise) an exception in Python. Syntax :- raise ExceptionType("Custom error message")

```python
age=int(input("Enter your age:"))
if age<0:
    raise ValueError("Age cannot be negative.")
else:
    print("Valid age Entered:",age)
```

```python
try:
    salary=float(input("Enter your salary:"))
    if salary<0:
        raise ValueError("Salary cannot be negative.")
    else:
        print("Salary recorded:",salary)
except ValueError as e:
    print("Error:",e)
```

```python
# Bank-Money transfer module
amount=int(input("Enter the amount to transfer:"))
balance=int(input("Enter your balance:"))
acc_bal=0
def money_send(amount,balance):
    if amount<0:
        raise Exception("amount cannot be -ve")
    if balance<amount:
        raise Exception("You dont have sufficient balance")
    else:
        acc_bal=balance-amount
        print("Money is transferred successfully and your current balance is:",acc_bal)
try:
    money_send(amount,balance)
finally:
    print("transaction is successful")        
```

### Custom Exception

```python
#raise error
class NegativeAgeError(Exception):
    pass
age=int(input("Enter your age:"))
if age<0:
    raise NegativeAgeError("Age cannot be negative!")
else:
    print("Valid age:",age)     
```

```python
# Handling error using try
class NegativeAgeError(Exception):
    pass
try:
    age=int(input("Enter your age:"))
    if age<0:
        raise NegativeAgeError("Age should not be negative")
        print("Age is valid:",age)
except NegativeAgeError as e:
    print("Error:",e)   
```

```python
# Custom Exception in business Logic
class LowBalanceError(Exception):
    pass
balance=int(input("Enter your bank balance:"))
try:
    if balance<500:
        raise LowBalanceError("Balance is below minimum request")
    else:
        print("transaction Successful!")
except LowBalanceError as e:
    print("Transaction failed:",e)
```
