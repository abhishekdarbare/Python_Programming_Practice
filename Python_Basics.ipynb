```python
num1 = 5
num2 = 5
print(id(num1))
print(id(num2))

print(num1==num2)
print(num1 is num2) # identity (same object in memory)

# Python has an optimisation technique called integer interning (or caching).
# For small integers (from -5 to 256), Python pre-creates these objects at startup and reuses them throughout the program.

a = 1000
b = 1000
print(id(a))
print(id(b))
print(a is b)   # For 1000, Python doesn’t cache it (outside the -5 to 256 range), so False
print(a == b) 

a=54
b=63
print(id(a))
print(id(b))
print(a is b)  #identity (different object in memory)
```

```python
T1 = (1, 2, 3)
T2 = (1, 2, 3)

print(T1 is T2) # Compare memory location-wise,i.e address value
print(T1 == T2) # Compare element wise

lst1 = [1, 2, 3]
lst2 = [1, 2, 3]
lst3 = lst1     # Since we have used the assignment operator, the memory location of lst1 will be equal to lst3

print(id(lst1))
print(id(lst2))
print(id(lst3))

print(lst1 is lst2)
print(lst1 is lst3)
print(lst1 == lst2)


str1 = "hello world"
str2 = "hello world"

print(str1 is str2)
```
