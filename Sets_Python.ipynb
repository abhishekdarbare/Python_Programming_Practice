### *Sets*

#### A set is an unordered collection of items.
#### Every set element is unique (no duplicates) and must be immutable (cannot be changed).
#### However, a set itself is mutable. We can add or remove items from it.
#### Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.
#### Characteristics:
##### 1) Unordered
##### 2) Mutable
##### 3) No Duplicates
##### 4) Can't contain mutable data types --> i.e. set,List so 2D set not possible

```python
# empty
#s={}  # <class 'dict'>
s=set()
print(s)
print(type(s))
```

```python
# 1D and 2D
s1 = {1,2,3}
print(s1)
s2 = {1,2,3,{4,5}} #Error: unhashable type: 'set'
# Python sets can only contain hashable (immutable) elements like a tuple(4,5).
#  A set like {4, 5} or say[4,5] is mutable, so it cannot be nested directly inside another set.
print(s2)
```

```python
# set can't have mutable items
s6 = {1,2,[3,4]}
print(s6) #Error: unhashable type: 'list'
```

```python
# homo and hetro
s3 = {1,'hello',4.5,(1,2,3)} #tuple can be used as nested directly as tuple is immutable
print(s3)
```

```python
# using type conversion
s4 = set([1,2,3.1])
s5 = set("hello")
print(s4)
print(type(s4))
print(s5)
print(type(s5))
```

```python
# duplicates not allowed
s5 = {1,1,2,2,3,3}
print(s5)
```

```python
# IMP quest
s1 = {1,2,3,2}
s2 = {3,2,1}
print(s1 == s2)
print(s1 is s2)
```

#### Accessing Items --> Not possible in sets(as set is unordered collection)

```python
s1 = {1,2,3,4}
print(s1[2])  # TypeError: 'set' object is not subscriptable
print(s1[0:3]) # TypeError: 'set' object is not subscriptable
# sets are unordered collections of unique elements
```

#### Editing Items --> Not possible

```python
s1 = {1,2,3,4}
s1[0] = 100 #TypeError: 'set' object does not support item assignment
print(s1)
```

#### Adding Items -> add(),update()
#### add() -> It adds the 1 item in set.
#### update() -> By using this function we can Adds multiple elements from an iterable (list, tuple, set)

```python
S = {1,2,3,4,5} 
#S.add((5,6,7,8))
S.add("Pune")
print(S)

S1= {1,2,3,4,5} 
S1.update([5,6,7,"yes",1])
print(S1)
```

#### Deleting Items -> discard(),remove(),clear(),del()
##### discard() -> It is used to delete the element from set.If element is not present in set then it will not throws any error.
##### remove() -> It is used to delete the element from set.If element is not present in set then it will  throws an error.
##### pop --> It deletes random element from set.
##### clear() -> This function removes all elements from set and make it empty.
##### del() -> This function is used to delete the entire set from memory.

```python
s={1,2,3,4,5}
```

```python
# discard
s.discard(4)
# s.discard(7) 
print(s)
```

```python
# remove
s.remove(7)  #Error as 7 is not present in set
print(s)
```

```python
# pop --> It deletes random element from set.
print(s)
s.pop()
print(s)
```

```python
# clear--> It deletes all elements from set and make it empty.
s = {1,2,3,4,5}
s.clear()
print(s)
```

```python
# del
s = {1,2,3,4,5}

print(s)
#del s[0]
#print(s)

del s
print(s) #Error as set is already deleted and set does not exists
```

### Set Functions
#### len/sum/min/max/sorted

```python
s = {3,1,4,5,2,7,7,2}
print(len(s))
print(sum(s))
print(min(s))
print(max(s))
print(sorted(s,reverse=True))
```

### isdisjoint -->Returns True if both sets having different elements

```python
s1 = {1,2,3,4,7}
s2 = {11,8,5,6,4}
print(s1.isdisjoint(s2))
```

### issubset/issuperset

```python
s1 = {1,2,3,4,5}
s2 = {3,4,5}
print(s2.issubset(s1))
print(s1.issuperset(s2))
```

#### union(|)                          Combines two sets (no duplicates)       s1.union(s2) or s1 | s2       
#### intersection(set2) (&)            Common elements in both sets            s1.intersection(s2) or s1 & s2             
#### difference(set2)  (-)             Elements in s1 but not in s2            s1.difference(s2) or s1 - s2                
#### symmetric_difference(set2)(^)     gives uncommon elements from both set   s1.symmetric_difference(s2) or s1 ^ s2      

```python
# Set Operation
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
```

```python
# Union(|) -> It ignore the duplicates
print(s1 | s2)
s1.union(s2)
```

```python
# Intersection(&)
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

print(s1 & s2)
s1.intersection(s2)
```

```python
# Difference(-) --> s1 - s2 i.e. items in s1 but not in s2
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

print(s2 - s1)
s2.difference(s1)
print(s1 - s2)
```

```python
# Symmetric Difference(^)--> gives uncommon elements from both set
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

print(s1 ^ s2)
s1.symmetric_difference(s2)
symt_diff=s1.difference(s2).union(s2.difference(s1))
print(symt_diff)
```

### Set Comprehension

```python
# print a set containing the items from 1 to 10
print({i for i in range(1,11)})
```

```python
# print a set containing even the items from 1 to 10
print({i for i in range(1,11) if i%2==0})
```

```python
# print a set containing squares of only items greater than 5 from 1 to 10
print({i**2 for i in range(1,11) if i>5})
```
